```python
from keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
import numpy as np

class NearestNeighbor(object):
    def __init__(self):
        pass 
    
    def train(self, x, y):
        self.x = x.reshape(x.shape[0], -1)
        self.y = y
        
    def predict(self, x):
        # initialize y_pred = [0, ..., ]
        num_test = x.shape[0]
        print('total test number:', num_test)
        
        # flatten
        x = x.reshape(num_test, -1)
        y_pred = np.zeros(num_test, dtype=self.y.dtype)
        
        for i in range(num_test):
            distances = np.sum(np.abs(self.x - x[i, :]), axis=1)
            min_index = np.argmin(distances)
            y_pred[i] = self.y[min_index]
            print(i, y_pred[i])
        return y_pred 
```

```python
%%time
nn = NearestNeighbor()
nn.train(x_train, y_train)
y_pred = nn.predict(x_test)
print('accuracy: %f', np.mean(y_pred == y_test))
```
